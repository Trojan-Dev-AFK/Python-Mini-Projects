```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def get_cosine_similarity(text1, text2):
    embeddings1 = model.encode(text1, convert_to_tensor=True)
    embeddings2 = model.encode(text2, convert_to_tensor=True)
    
    similarity = util.pytorch_cos_sim(embeddings1, embeddings2)
    
    return similarity.item()
```

```python
def compare_csv(filename, column1, column2):
    df = pd.read_csv(filename)
    
    for index, row in df.iterrows():
        cell1 = row[column1]
        cell2 = row[column2]

        similarity = get_cosine_similarity(cell1, cell2)
        print(f"Similarity between '{cell1}' and '{cell2}': {similarity:.2f}")
```

```python
csv_file = "./Data/ComparisonData.csv"
column_to_compare1 = "comment"
column_to_compare2 = "compare_comment"
    
compare_csv(csv_file, column_to_compare1, column_to_compare2)
```
